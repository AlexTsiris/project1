```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns  
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/AlexTsiris/project1/main/business_dynamics.csv"
business_df = pd.read_csv(url)
```

```python
business_df.columns
```

```python
business_df.shape
```

```python
alaska_data = business_df[business_df["State"] == "Alaska"]
```

```python
plt.figure(figsize=(4, 3)) 
sns.scatterplot(data=alaska_data, x='Year', y='Data.Job Creation.Births')
plt.title('Job Creation Births in Alaska')
plt.xlabel('Year')
plt.ylabel('Job Creation Births')
```

```python
fig, ax = plt.subplots (figsize=(4, 3),  ncols=2)
plt.subplots_adjust(wspace = 2)
sns.boxplot(y=alaska_data['Data.Job Creation.Births'], ax = ax[0])
sns.boxplot(y=alaska_data['Data.Job Destruction.Deaths'], ax = ax[1])
#plt.title('Job Creation Births in Alaska')
#plt.ylabel('Job Creation Births')
```

```python
fig, ax = plt.subplots(2, figsize = (10,7))
plt.subplots_adjust(wspace = 1)

sns.barplot(data=alaska_data, x='Year', y='Data.Job Creation.Births', ax = ax[0])
sns.barplot(data=alaska_data, x='Year', y='Data.Job Destruction.Deaths', ax = ax[1])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Группировка данных по 5-летним интервалам и вычисление суммы значений
alaska_data['Year_Group'] = pd.cut(alaska_data['Year'], bins=range(1980, 2026, 5), right=False)
grouped_data = alaska_data.groupby('Year_Group').sum().reset_index()

# Построение гистограмм
fig, ax = plt.subplots(2, figsize=(10, 7))
plt.subplots_adjust(hspace=0.5)

sns.barplot(data=grouped_data, x='Year_Group', y='Data.Job Creation.Births', ax=ax[0])
ax[0].set_title('Distribution of Job Creation Births by 5-year interval')
ax[0].set_ylabel('Count')
ax[0].set_xlabel('Year Interval')

sns.barplot(data=grouped_data, x='Year_Group', y='Data.Job Destruction.Deaths', ax=ax[1])
ax[1].set_ylabel('Count')
ax[1].set_xlabel('Year Interval')

plt.xticks(rotation=45)  # Повернуть подписи по оси X
plt.show()
```
